```python
# Let us calculate BMI(Body Mass Index) for these data using Python List
height = [1.78, 1.63, 1.55] # Python List
weight = [70.4, 64.7, 58.9] # Python List

#Now calculating BMI - Formula is = (weight / height) power of 2
(weight / height) ** 2

# Python has no idea on how to do calculations on a list
```

```python
# Let us calculate BMI(Body Mass Index) for these data using Numpy now
height = [1.78, 1.63, 1.55]
np_height = np.array(height) # Numpy Array

weight = [70.4, 64.7, 58.7]
np_weight = np.array(weight) # Numpy Array

#Now calculating BMI - Formula is = (weight / height) power of 2
BMI = (np_weight / np_height) ** 2
BMI
# Numpy Array provides elegant solution
```

```python
# How Python List and Numpy Arrays are different from each other
python_list = [1,2,3]
np_array = np.array([1,2,3])

# How '+' operator functions for these
print ('Python List: ', python_list + python_list) # appends the list
print('Numpy Array: ',np_array + np_array) # performs addition
```

```python
# Numpy Subsetting
BMI[1] #1574.
BMI > 1500 # array([ True,  True, False])
BMI[BMI > 1560] # array([1564.24693852, 1575.55421732])
```

```python
# Some mathematical operations on Numpy Array
print('Mean: ', np.mean(BMI[:])) # Mean
print('Median: ', np.median(BMI[:])) # Median
print('Corr Coefficient:\n', np.corrcoef(BMI[:], BMI[:])) # Corrctive Co-efficient
print('Standard Deviation: ', np.std(BMI[:])) # Standard Deviation
```

```python
mu, sigma =0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
```

```python
abs(mu - np.mean(s)) < 0.01
abs(sigma - np.std(s, ddof=1)) < 0.01
```

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s, 30, normed=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
                  np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
             linewidth=2, color='r')
plt.show()
```

```python
# Array Procesing
import matplotlib.pyplot as plt
points = np.arange(-5, 5, 0.1)
dx, dy = np.meshgrid(points, points)
z = np.sin(dx) + np.sin(dy)
plt.plot(z)
plt.title('Plot for sin(dx) + sin(dy)')
plt.show()
```
